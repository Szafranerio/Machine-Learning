```python
user_over_200 = df['user'].value_counts()
user_over_200 = user_over_200[user_over_200 > 200].index
```

```python
isbn_over_100 = df['isbn'].value_counts()
isbn_over_100 = isbn_over_100[isbn_over_100 > 100].index
```

```python
ready_df = df.loc[df['user'].isin(user_over_200) & df['isbn'].isin(isbn_over_100)]
ready_df.head()
```

```python
ready_df[['title', 'user']].duplicated().value_counts()
```

```python
ready_df = ready_df.drop_duplicates(['user', 'title'])
ready_df.head()
```

```python
ready_df_p = ready_df.pivot_table(index= 'title', columns='user', values='rating').fillna(0)
ready_df_p
```

```python
from scipy.sparse import csr_matrix
```

```python
df_knn = csr_matrix(ready_df_p.values)
```

Machine Learing calssification

```python
from sklearn.neighbors import NearestNeighbors
```

```python
knn = NearestNeighbors(metric='cosine', algorithm='brute')
knn.fit(df_knn)
```

```python
def get_recommends(book=""):
    recommended_books = [book, []]
    
    distances, indices = knn.kneighbors([ready_df_p.loc[book]], 6, return_distance=True)
    
    recommended_indices = ready_df_p.iloc[np.flip(indices[0])[:-1]].index.to_list()
    recommended_distances = list(np.flip(distances[0])[:-1])
    
    for i in zip(recommended_indices, recommended_distances):
        recommended_books[1].append(list(i))

    return recommended_books
```

```python
books = get_recommends("Where the Heart Is (Oprah's Book Club (Paperback))")
print(books)

def test_book_recommendation():
  test_pass = True
  recommends = get_recommends("Where the Heart Is (Oprah's Book Club (Paperback))")
  if recommends[0] != "Where the Heart Is (Oprah's Book Club (Paperback))":
    test_pass = False
  recommended_books = ["I'll Be Seeing You", 'The Weight of Water', 'The Surgeon', 'I Know This Much Is True']
  recommended_books_dist = [0.8, 0.77, 0.77, 0.77]
  for i in range(2): 
    if recommends[1][i][0] not in recommended_books:
      test_pass = False
    if abs(recommends[1][i][1] - recommended_books_dist[i]) >= 0.05:
      test_pass = False
  if test_pass:
    print("You passed the challenge! 🎉🎉🎉🎉🎉")
  else:
    print("You haven't passed yet. Keep trying!")

test_book_recommendation()
```

```python
print('Enter the book name to find recommendations: ')
x = input("")

recommended_books = get_recommends(x)

recommended_books_list = [f"{book} (distance: {dist:.2f})" for book, dist in recommended_books[1]]

print('Recommended books for ' + x + ': ' + ', '.join(recommended_books_list))
```
