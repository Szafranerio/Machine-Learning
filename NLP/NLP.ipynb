# üß† Tokenization in Python

**Tokenization** is the process of splitting text into smaller units called **tokens**.  
These can be:
- Words
- Sentences
- Characters
- Subwords (used in transformer models)

# üî† Unigrams, Bigrams, Trigrams in Python

## üìå What Are They?

**N-grams** are continuous sequences of `n` items (usually words) from a given text.

- **Unigrams (n=1)**: Single words  
  Example: `["I", "love", "Python"]`

- **Bigrams (n=2)**: Pairs of consecutive words  
  Example: `[("I", "love"), ("love", "Python")]`

- **Trigrams (n=3)**: Triplets of consecutive words  
  Example: `[("I", "love", "Python")]`

N-grams help capture context and are useful in tasks like:
- Text prediction
- Language modeling
- Text classification

## üîπ Generate N-grams using NLTK

```python
from nltk.tokenize import word_tokenize
from nltk.util import ngrams

text = "I love learning Python"
tokens = word_tokenize(text)

unigrams = list(ngrams(tokens, 1))   # [('I',), ('love',), ('learning',), ('Python',)]
bigrams = list(ngrams(tokens, 2))    # [('I', 'love'), ('love', 'learning'), ('learning', 'Python')]
trigrams = list(ngrams(tokens, 3))   # [('I', 'love', 'learning'), ('love', 'learning', 'Python')]

# üåø Lemmatization in Python

## üìò What is Lemmatization?

**Lemmatization** is the process of converting a word to its **base or dictionary form** (*lemma*), considering the context and part of speech.

- More **linguistically accurate** than stemming
- Returns **real words**
- Requires **POS tags** for best results

## üîπ Stemming

- **Definition**: Reduces words to their **root/base form** by chopping off prefixes or suffixes.
- **Doesn't guarantee valid words**
- **Fast** but sometimes **inaccurate**

````python
#Lemmatization
from nltk.stem import WordNetLemmatizer
lemmitizer = WordNetLemmatizer()

lemmitizer.lemmatize('machines')
````

````python
#Stemming
from nltk.stem.porter import PorterStemmer
porter = PorterStemmer()
porter.stem('running')
````

# üßÆ What is CountVectorizer and Why Use It?

## üìò What is CountVectorizer?

`CountVectorizer` is a tool from **scikit-learn** used to convert **text documents into numerical feature vectors** based on word frequency.  
This process is part of the **Bag of Words (BoW)** model in Natural Language Processing (NLP).

It transforms a collection of text documents into a matrix where:
- **Rows** represent documents
- **Columns** represent vocabulary words
- **Values** represent the **count** of each word in each document

---

## üõ†Ô∏è How It Works

1. **Tokenization**: Breaks text into individual words (tokens)
2. **Vocabulary Building**: Identifies unique words across all documents
3. **Vectorization**: Counts how often each word appears per document

````python
#Count vectorizer
from sklearn.feature_extraction.text import CountVectorizer
count_vectorize = CountVectorizer(max_features=500)
data_count = count_vectorize.fit_transform(groups.data)
data_count
````

````python
print(count_vectorize.get_feature_names_out())
````

````python
#Text processing
def is_letter_only(word):
    for char in word:
        if not char.isalpha():
            return False
    return True
data_cleaned = []
````

````python
for doc in groups.data:
    # Filter words that are letters only
    words = [word for word in doc.split() if is_letter_only(word)]
    doc_cleaned = ' '.join(words)  # Join back into a cleaned string
    data_cleaned.append(doc_cleaned)
````

# üåÄ What is t-SNE?

## üîç Definition

**t-SNE** (t-Distributed Stochastic Neighbor Embedding) is a **dimensionality reduction** technique used to visualize **high-dimensional data** in **2D or 3D space**.

---

## üéØ Purpose

- Helps to **visualize** complex datasets (e.g., text embeddings, image features).
- Preserves **local structure**: points that are close in high dimensions remain close in 2D/3D.
- Especially useful in **unsupervised learning** and **clustering**.

---

## üõ†Ô∏è How It Works (High-Level)

1. Computes **pairwise similarities** between points in high-dimensional space.
2. Maps those similarities into a **lower-dimensional space**.
3. Uses a **probabilistic approach** to keep similar points close and dissimilar ones apart.

---

## üìä Use Cases

- Visualizing word embeddings (e.g., Word2Vec, BERT)
- Exploring clusters from unsupervised models
- Understanding high-dimensional feature spaces

````python
count_vectorize = CountVectorizer(max_features=500)
data_count = count_vectorize.fit_transform(groups_3)
data_count
````

````python
all_names = set(names.words())
count_vector_sw = CountVectorizer(stop_words='english', max_features=500)
lemmitizer = WordNetLemmatizer()
data_cleaned = []

for doc in groups_3.data:
    doc = doc.lower()
    doc_cleaned = ' '.join(lemmitizer.lemmatize(word)
                           for word in doc.split()
                           if is_letter_only(word) and word not in all_names)
    data_cleaned.append(doc_cleaned)
data_cleaned_count = count_vector_sw.fit_transform(data_cleaned)
print(count_vector_sw.get_feature_names_out())
````
