```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
from sklearn.datasets import make_blobs
```

```python
data = make_blobs(n_samples=200, n_features=2, centers=4, cluster_std=1.8, random_state=101)
```

```python
data
```

```python
plt.scatter(data[0][:,0], data[0][:,1], c=data[1], cmap='rainbow')
plt.show()
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=4)
kmeans.fit(data[0])
```

```python
print('Cluster membership: ', kmeans.labels_)
```

```python
print('Clusters Centers:', kmeans.cluster_centers_)
```

```python
fig, (ax1, ax2) = plt.subplots(1,2, sharey=True, figsize=(10,6))

ax1.set_title('K Means')
ax1.scatter (data[0][:,0], data[0][:,1], c= kmeans.labels_)

ax2.set_title('Original')
ax2.scatter (data[0][:,0], data[0][:,1], c= data[1])
plt.show()
```

Failure cases of k-Means

```python
from sklearn.datasets import make_moons
X, y = make_moons(n_samples=1000, noise=0.05, random_state=0)
```

```python
kmeans = KMeans(n_clusters=2)
kmeans.fit(X)
y_pred = kmeans.predict(X)
```

```python
#Not accurate clustering 
plt.scatter(X[:,0],X[:,1], c=y_pred, s=60)
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1])
plt.show()
```
