```python
cat_feats = ['purpose']
```

```python
final_data = pd.get_dummies(data, columns=['purpose'], drop_first=True)
```

```python
final_data.info()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = final_data.drop('not.fully.paid',axis=1)
y = final_data['not.fully.paid']
X_train, X_test, y_train, y_test = train_test_split (X, y, test_size=0.3, random_state=3)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dtree = DecisionTreeClassifier()
```

```python
dtree.fit(X_train, y_train)
```

```python
predictions = dtree.predict(X_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(confusion_matrix(y_test,predictions))
print(classification_report(y_test,predictions))
```

Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfor = RandomForestClassifier(n_estimators=2000)
```

```python

rfor.fit(X_train, y_train)
```

```python
rfor_predict = rfor.predict(X_test)
```

```python
print(confusion_matrix(y_test,rfor_predict))
print(classification_report(y_test,rfor_predict))
```
